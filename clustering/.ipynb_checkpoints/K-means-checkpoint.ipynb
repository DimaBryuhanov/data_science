### K-means clustering
is an unsupervised algorithm with the known number of clusters, based on the number of centroids (K) and nearest mean values, given the Euclidean distance measurements between observations.

#### Use cases:
1. Market price / cost modeling
2. Insurance claim fraud detection
3. Hedge fund classification
4. Customer segmentation

#### Data preparation required:
1. Variables must be scaled
2. Define the number of centroids from a scatter plot

## Setting up

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
```

```python
%matplotlib inline
plt.figure(figsize=(7,4))
```

```python
iris = datasets.load_iris()

X = scale(iris.data)
y = pd.DataFrame(iris.target)

variable_names = iris.feature_names
X[:10]
```

## Building & running the model

```python
clustering = KMeans(n_clusters=3, random_state=5)
clustering.fit(X)
```

## Plotting the model outputs

```python
iris_df = pd.DataFrame(iris.data)
iris_df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
y.columns = ['targets']
```

```python
color_theme = np.array(['darkgrey', 'lightsalmon', 'powderblue'])

plt.subplot(1,2,1)
plt.scatter(x=iris_df.petal_length, y=iris_df.petal_width, c=color_theme[iris.target], s=50)
plt.title('Ground truth classification')

plt.subplot(1,2,2)
plt.scatter(x=iris_df.petal_length, y=iris_df.petal_width, c=color_theme[clustering.labels_], s=50)
plt.title('K-means prediction')
```

```python
relable = np.choose(clustering.labels_, [1, 0, 2]).astype(np.int64)

plt.subplot(1,2,1)
plt.scatter(x=iris_df.petal_length, y=iris_df.petal_width, c=color_theme[iris.target], s=50)
plt.title('Ground truth classification')

plt.subplot(1,2,2)
plt.scatter(x=iris_df.petal_length, y=iris_df.petal_width, c=color_theme[relable], s=50)
plt.title('K-means prediction')
```

## Evaluating the clustering results

```python
print(classification_report(y, relable))
```
